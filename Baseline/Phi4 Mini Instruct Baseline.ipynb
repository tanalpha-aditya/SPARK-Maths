```python
SYSTEM_PROMPT = '''
You are a math expert. Users will provide you with a problem statement and you are required to solve 
the question by providing proper reasoning.

Please provide output strictly in following format. Every content should be in one of the tabs

Respond in the following format:
<think>
<<reason for answer strictly in 100 words>>
</think>
<answer>
<<strictly integral answer>>
</answer>
'''

USER_PROMPT = '''
Question: {0}'''
```

```python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

endpoint = "https://srsinha-ai-service.services.ai.azure.com/models"
model_name = "Phi-4-mini-instruct"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(""),
#     response_format = 'json_object'
)
```

```python
dataset_path = '~/Desktop/project/SPARK-Maths/Dataset/MATH/Pre_processed_test/filtered_test_serial.csv'
```

```python
import pandas as pd
```

```python
df = pd.read_csv(dataset_path)
```

```python
df.info()
```

```python
#def get_llm_response():
from tqdm.notebook import tqdm
all_answers = {}
failed_idx = []
for idx,problem in tqdm(df.iterrows()):
    print(problem['question'])
    try:
        response = client.complete(
            messages=[
                SystemMessage(content=SYSTEM_PROMPT),
                UserMessage(content=USER_PROMPT.format(problem['question']))
            ],
            max_tokens=300,
            temperature=0.8,
            top_p=0.1,
            presence_penalty=0.0,
            frequency_penalty=0.0,
            model=model_name
        )
        print(response.choices[0].message.content)
        answer = float(response.choices[0].message.content.split('<answer>')[-1].split('</answer>')[0][1:-1])
        print(f'For {idx} got answer {answer}')
        all_answers[idx] = answer
    except Exception as e:
        print(e)
        print('{0} failed'.format(idx))
        failed_idx.append(idx)
        all_answers[idx] = -1
```

```python
import re

def extract_answers_by_index(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        content = file.read()

    # Split based on lines ending in either 'failed' or 'got answer'
    question_blocks = re.split(r'\n(?:\d+ failed|For \d+ got answer.*?)\n', content)

    answers = {}
    for idx, block in enumerate(question_blocks):
        match = re.search(r'<answer>\s*(.*?)\s*</answer>', block, re.DOTALL)
        if match:
            answers[idx] = match.group(1).strip()

    return answers

# Usage
file_path = "Gsm8k_baseline_report.txt"
answers_dict = extract_answers_by_index(file_path)

# Example preview
for idx in range(min(5, len(answers_dict))):
    print(f"{idx}: {answers_dict[idx]}\n")
```
